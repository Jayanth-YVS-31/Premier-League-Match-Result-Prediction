##### Now we encode the Target column i.e FTR(Full Time Result)
Home(H) : 1

Draw(D) : 0

Away(A) : -1

```python
data['FTR'] = data['FTR'].map({'H':1, 'D':0, 'A':-1})
```

Creating a column as `GoalDifference` with HomeGoals - AwayGoals

```python
data['GoalDifference'] = data['HomeGoals'] - data['AwayGoals']
```

```python
data.drop(columns = ['FTR_num'], inplace = True)
data.head(3)
```

We are considering the teams previous preformance of goals scoring, so that we can have a understanding of how are they performing. So, we use rolling average on last 5 matches.

```python
data['HomeGoalsAvg_5'] = data.groupby('Home')['HomeGoals'].rolling(5).mean().reset_index(0, drop=True)
data['AwayGoalsAvg_5'] = data.groupby('Away')['AwayGoals'].rolling(5).mean().reset_index(0, drop=True)
```

```python
data.tail(2)
```

##### Finding the Team form features
Football is all about mentality, so let's find the streak they are in, whether it is wins or losses or both.
###### Finding the streaks in the data

```python
#  HomeWin and AwayWin based on FTR column to calculate the streaks
data['HomeWin'] = (data['FTR'] == 1).astype(int)
data['AwayWin'] = (data['FTR'] == -1).astype(int)
```

```python
data['HomeWinStreak_5'] = 0
data['AwayWinStreak_5'] = 0

for team in data['Home'].unique():
    team_home_games = data[data['Home'] == team]
    
    data.loc[data['Home'] == team, 'HomeWinStreak_5'] = team_home_games['HomeWin'].rolling(5, min_periods=1).sum().values

for team in data['Away'].unique():
    team_away_games = data[data['Away'] == team]
    
    data.loc[data['Away'] == team, 'AwayWinStreak_5'] = team_away_games['AwayWin'].rolling(5, min_periods=1).sum().values
```

```python
data['HomeWinStreak_5'].nunique()
```

```python
# removing the HomeWin and AwayWin
data.drop(columns = ['HomeWin', 'AwayWin'], inplace = True)
```

```python
data.tail(5)
```

Adding the last match result in the data

```python
data['LastHomeResult'] = data.groupby('Home')['FTR'].shift(1)
data['LastAwayResult'] = data.groupby('Away')['FTR'].shift(1)
data.tail(3)
```

It might cause disturbance in the prediction.

Now we'll make the dataset symmetric to address the first problem

"Creating a symmetric dataset by including both "home" and "away" perspectives for each match improves the model's generalization, balances class distributions, and captures reverse dynamics. This approach gives the model more examples and helps it recognize patterns from both sides, enhancing its predictive accuracy for future games."

```python
data.shape
```

```python
data_reverse = data.copy()
data_reverse.rename(columns={'Home': 'Away', 'Away': 'Home', 'HomeGoals': 'AwayGoals', 'AwayGoals': 'HomeGoals', 'HomePoints': 'AwayPoints', 'AwayPoints': 'HomePoints'}, inplace=True)

# Update the reversed copy’s outcome to match the reverse perspective
data_reverse['FTR'] = data_reverse['FTR'] * -1  # Reverse outcome
data_reverse['GoalDifference'] = data_reverse['GoalDifference'] * -1  # Reverse goal difference

# Concatenate original and reversed datasets
data_symmetric = pd.concat([data, data_reverse], ignore_index=True)
```

#### Model Training

Now we have our data in a form that we can send it to a model. The question is what model we have to use? It's obvious that we have a multiclass classification.

I have an idea which i read online that we can use a meta-model to increase the accuracy of model. A Meta-model means using model on the data that is predicted by another model. Basically it is stacking models and send the data. 

what i want to do is to use XGBoost and RandomForest as a base model and the predictions from these are sent to another model like Logistic Regression. This process will give the benefits form different models and covers each others.

```python
test_x['DaysSinceLastAwayMatch'] = test_x['DaysSinceLastAwayMatch'].fillna(test_x['DaysSinceLastAwayMatch'].median())
```

```python
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier

# Initialize models
xgb = CatBoostClassifier()
rf = RandomForestClassifier()

# Train both models on the same training data
xgb.fit(train_x_3, train_y_3)
rf.fit(train_x_3, train_y_3)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

xgb_train_pred_label = np.argmax(xgb.predict_proba(train_x_3), axis=1)
rf_train_pred_label = np.argmax(rf.predict_proba(train_x_3), axis=1)

print("XGB model accuracy (train):", accuracy_score(train_y_3, xgb_train_pred_label))
print("RandomForest model accuracy (train):", accuracy_score(train_y_3, rf_train_pred_label))

xgb_train_pred = xgb.predict_proba(train_x_3)
rf_train_pred = rf.predict_proba(train_x_3)
stacked_train_x = np.hstack((xgb_train_pred, rf_train_pred))

xgb_test_pred_label = np.argmax(xgb.predict_proba(test_x_3), axis=1)
rf_test_pred_label = np.argmax(rf.predict_proba(test_x_3), axis=1)

print("XGB model accuracy (test):", accuracy_score(test_y_3, xgb_test_pred_label))
print("RandomForest model accuracy (test):", accuracy_score(test_y_3, rf_test_pred_label))

xgb_test_pred = xgb.predict_proba(test_x_3)
rf_test_pred = rf.predict_proba(test_x_3)
stacked_test_x_3 = np.hstack((xgb_test_pred, rf_test_pred))
```

```python
from sklearn.linear_model import LogisticRegression

meta_model = LogisticRegression()
meta_model.fit(stacked_train_x, train_y_3)
```

```python
final_predictions = meta_model.predict(stacked_test_x_3)

# Optionally, evaluate performance
from sklearn.metrics import accuracy_score

print("Ensemble model accuracy:", accuracy_score(test_y_3, final_predictions))
import joblib
joblib.dump(meta_model, 'meta_model_3.pkl')
```

#### Hyperparameter Tuning

```python
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
import numpy as np

# Define parameter grids for each model
xgb_params = {'n_estimators': [100, 200, 300], 'learning_rate': [0.01, 0.1, 0.2], 'max_depth': [3, 5, 7], 'subsample': [0.7, 0.8, 1.0]}
catboost_params = {'iterations': [100, 200, 300], 'learning_rate': [0.01, 0.1, 0.2], 'depth': [4, 6, 8]}
rf_params = {'n_estimators': [100, 200, 300], 'max_depth': [None, 10, 20, 30], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4], 'bootstrap': [True, False]}
meta_params = {'C': [0.01, 0.1, 1, 10, 100], 'solver': ['lbfgs', 'liblinear'], 'penalty': ['l2']}

# Hyperparameter tuning for each model
xgb_random_search = RandomizedSearchCV(XGBClassifier(), xgb_params, cv=3, n_iter=10, scoring='accuracy', n_jobs=-1, random_state=42)
xgb_random_search.fit(train_x_3, train_y_3)
best_xgb = xgb_random_search.best_estimator_

catboost_random_search = RandomizedSearchCV(CatBoostClassifier(verbose=0), catboost_params, cv=3, n_iter=10, scoring='accuracy', n_jobs=-1, random_state=42)
catboost_random_search.fit(train_x_3, train_y_3)
best_catboost = catboost_random_search.best_estimator_

rf_random_search = RandomizedSearchCV(RandomForestClassifier(), rf_params, cv=3, n_iter=10, scoring='accuracy', n_jobs=-1, random_state=42)
rf_random_search.fit(train_x_3, train_y_3)
best_rf = rf_random_search.best_estimator_

# Fit tuned models
best_xgb.fit(train_x_3, train_y_3)
best_rf.fit(train_x_3, train_y_3)
best_catboost.fit(train_x_3, train_y_3)

# Get predictions for stacking
xgb_train_pred = best_xgb.predict_proba(train_x_3)
rf_train_pred = best_rf.predict_proba(train_x_3)
catboost_train_pred = best_catboost.predict_proba(train_x_3)
stacked_train_x = np.hstack((xgb_train_pred, rf_train_pred, catboost_train_pred))

xgb_test_pred = best_xgb.predict_proba(test_x_3)
rf_test_pred = best_rf.predict_proba(test_x_3)
catboost_test_pred = best_catboost.predict_proba(test_x_3)
stacked_test_x_3 = np.hstack((xgb_test_pred, rf_test_pred, catboost_test_pred))

# Tune meta-model
meta_random_search = RandomizedSearchCV(LogisticRegression(), meta_params, cv=3, n_iter=10, scoring='accuracy', n_jobs=-1, random_state=42)
meta_random_search.fit(stacked_train_x, train_y_3)
best_meta_model = meta_random_search.best_estimator_

# Fit and evaluate the final ensemble model
best_meta_model.fit(stacked_train_x, train_y_3)
final_predictions = best_meta_model.predict(stacked_test_x_3)
print("Tuned Ensemble model accuracy:", accuracy_score(test_y_3, final_predictions))
joblib.dump(best_meta_model, 'tuned_meta_model_3.pkl')
```
