### Imporving precision with Rolling averages

Rolling avarages is a method which uses the previous form of the team to predict the result of the present. Example: If a team won 2 in a row, everyone expects them to win their 3 game, similary vicevera. But real life is a unpredictable and we can't gaurentee everything that we have in out paper but it still can provide us some information which can prepare us to excpet something.

```python
grouped_matches = data.groupby('team')
```

```python
mancity = grouped_matches.get_group('Manchester City')
```

```python
mancity
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed = 'left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
rolling_averages(mancity, cols, new_cols) #trying for only mancity
```

```python
# now for the whole data
matches_rolling = data.groupby('team').apply(lambda i : rolling_averages(i, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_pred(data, predictors):
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] > '2022-01-01']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test['target'], prediction = preds), index = test.index)
    return combined, preds
    
```

```python
combined, preds = make_pred(matches_rolling, predictors + new_cols)
```

```python
combined = combined.merge(matches_rolling[['date', 'team', 'opponent', 'result']], left_index = True, right_index = True)
```

```python
combined
```

Now we have data and some predictions whether they are right or wrong. But there is a problem, here we are seeing two predictions for a single match Example : For Arsenal vs Burnley we got a prediction and for the same match we got a prediction somewhere down the dataframe. This is obviously not good for making predictions.

```python
acc = accuracy_score(test['target'], preds)
acc
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key : key

map_values = {
    'Brighton and Hove Albion' : 'Brighton',
    'Manchester United' : 'Manchester Utd',
    'Newcastle United' : 'Newcastle Utd',
    'Tottenham Hotspur' : 'Tottenham',
    'West Ham United' : 'West Ham',
    'Wolverhampton Wanderers' : 'Wolves'
}

mapping = MissingDict(**map_values)
```

```python
combined['new_team'] = combined['team'].map(mapping)
```

```python
merged = combined.merge(combined, left_on = ['date', 'new_team'], right_on = ['date', 'opponent'])
```
